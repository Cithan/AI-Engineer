## Infinite Tree

![tree](https://storage.googleapis.com/rg-ai-bootcamp/internship/tree.png)

This is a special tree that keep growing, the growth rule is as follows:
- On the first level, the height of the tree is `1000`
- On the second level, the tree branch-out to `2` branches, each branch's height is `2/3` of the previous branch. Moreoever, the direction of the branched are rotated by `x` degree w.r.t the previous branch direction. (see the picture above)
- On the next level, the same process is repeated.

What is the total height of the tree on the `n`-th level?

The height is vertical height, not the length of the branch.

`height(1, 30) == 1000`

`height(2, 30) == 1000 + 2/3 * 1000 * cos(30)`

`0 > rotation_degree < 90`

````python
from math import cos, radians
def height(level, rotation_degree):
    # Implement your code her
    b= 0
    for x in range(1,level):
      a = (2/3)**(x) * 1000 * math.cos(math.radians(rotation_degree))
      b =  b + abs(a)
      c = 1000 + b
    return c

print(height(3, 10))
````

## PIN

![pin](https://storage.googleapis.com/rg-ai-bootcamp/internship/pin.png)

If you use Android, one way to unlock you phone is by swiping a pattern on the screen. The pattern is a sequence of 9 dots, and you can swipe horizontally, vertically, or diagonally.

You can't press the same dot twice.

For example, below are valid patterns with length == 3:

![valid-pin](https://storage.googleapis.com/rg-ai-bootcamp/internship/valid-pin.png)

To make the problem easier, we **only allow** swiping to the **adjacent dots** (horizontally, vertically, and diagonally). For example, the pattern below is invalid because they are not adjacent:

![invalid-pin](https://storage.googleapis.com/rg-ai-bootcamp/internship/invalid-pin.png)

The swiping can start from any dot, and the length of the pattern can be any number from 1 to 9.

### Challenge

Given the length of the pattern, return the number of possible pin patterns.

e.g.

`num_of_possible_pins(1) == 9`

`num_of_possible_pins(2) == 40`


````python
def is_valid_move(prev_dot, next_dot):
  ''' memeriksa pergerakan valid jika dot selanjutnya berdekatan dengan dot sebelumnya'''
  x1, y1 = prev_dot
  x2, y2 = next_dot
  return abs(x1 - x2) <= 1 and abs(y1 - y2) <=1 and (x1 != x2 or y1 != y2)

is_valid_move((1,1), (3,3))

def count_patterns(length, current_dot, visited):
  if length == 0:
    return 1 # mengembalikan nilai 1 jika panjang 0

  total_patterns = 0

  visited[current_dot[0]][current_dot[1]] = True # mark jika dot telah dimarking / visited
  # looping untuk penandaan jika dot telah di mark/visited
  for i in range(3):
    for j in range(3):
      next_dot = (i, j)
      if not visited[i][j] and is_valid_move(current_dot, next_dot):
        total_patterns += count_patterns(length - 1, next_dot, visited)

  # backtrack
  visited[current_dot[0]][current_dot[1]] = False

  return total_patterns

def num_of_possible_pins(n):
    # Your code here
    dots = [[False] * 3 for _ in range(3)]

    total_patterns = 0

    for i in range(3):
      for j in range(3):
        total_patterns += count_patterns(n - 1, (i, j), dots)
    return total_patterns

print(num_of_possible_pins(3))
````

## Secret Message

Doni and Dini likes to share message with each others. But they know that Dono always try to read their message. So they decide to encrypt their message.

To encrypt the message, Doni and Dini create a mapping table. e.g.

| From | To |
|------|----|
| A    | C  |
| B    | V  |
| C    | R  |
... and so on

When Doni or Dini wants to send a message, they will replace each character in the message with the corresponding character in the mapping table. For example, if they wants to send the word `ABC`, they will replace it with `CVR`.

And to decrypt the message, Doni or Dini will reverse the mapping table. e.g. `C` will be `A`, `V` will be `B`, and `R` will be `C`. So message `CVR` will be decrypted to `ABC`.

Dono knows about this, so he tries to decrypt the message. But he doesn't know the mapping table. Hmm, so how?

### Clue

Dono knows that the message is in Bahasa Indonesia. So he tries to find the most common character in Bahasa Indonesia. He found that the most common character is `A`, followed by `N`, `I`, `E`, `T`, and so on. So he assumes that the most common character in the message is `A`, and replace it with `A`. Then he tries to find the second most common character, and replace it with `N`. And so on.

Here is the frequency table of Bahasa Indonesia character Dono found:

```python
letter_freq = [('A', 19.64), ('N', 9.87), ('I', 8.28), ('E', 8.20), ('T', 5.4), ('M', 5.12), ('K', 4.8), ('S', 4.64), ('U', 4.56), ('D', 4.2), ('R', 4.04), ('O', 3.23), ('P', 3.1), ('B', 2.92), ('H', 2.76), ('G', 2.48), ('L', 2.4), ('Y', 1.88), ('J', 0.91), ('C', 0.64), ('F', 0.36), ('W', 0.24), ('V', 0.16), ('Z', 0.04), ('X', 0.02), ('Q', 0.01)

```

Of course the frequency of the character in the message will not be exactly the same as the frequency of the character in Bahasa Indonesia. But Dono assumes that it will be close enough. So, he just sort the character in the message by its frequency, and decrypt it with the character in Bahasa Indonesia with the same index.

So, `ZZZZBBBCC` will be decrypted to `AAAANNNII`.

### Tie

He found that _sometimes_ some characters in the message have the same frequency. For example, the message `ZZZABAB`. Both of the `A` and `B` have the same frequency. For such cases, he will decrypt it to `_` (to indicate that he doesn't know the character). So `ZZZABAB` will be decrypted to `AAA____`

### Non Alphabet Character

It turned out non alphabet character is NOT encrypted. So `ZZZ!@#$%^&*()90` can be decrypted to `AAA!@#$%^&*()90`.
All alphabets are in uppercase.
